3. Найдите у него наиболее часто встречающийся модуль суммы приходов/расходов. (\**)


#### Выберите все верные пункты:
    - 1) 1122957.89
    - 2) 15721.41
    - 3) 22459.16
    - 4) 13475494.63
    - 5) 107407.78
    - 6) 65019.26

##### Пояснения:
(\*) Если у клиента были транзакции [-10000, 10, 0, -10], то максимумом будет являться значение 10. <br>
(\**) Если у клиента были транзакции [-10000, 10, 0, -10], то наиболее встречающийся модуль суммы транзакций равен 10, и встретился он 2 раза.

````python
# Считываем данные и меняем индекс
data = pd.read_csv('data/transactions.csv', sep=',')
data = data.set_index('customer_id')

max_amount = data['amount'].max()
max_amount_index = data[data['amount'] == max_amount].index[0]
print(f"Максимальная трата: {max_amount}\nУ клиента: {max_amount_index}")

# Создаем список со всеми тратами, по модулю
arr_module = data.loc[max_amount_index]['amount'].abs().tolist()

# Создаем словарь с "сумма - количество раз"
from collections import Counter

count_dict = Counter(arr_module)

# Находим ключ с наибольшим значением в словаре
key_with_max_value = max(count_dict, key=lambda k: count_dict[k])

# Выводим ключ и его значение
print("Трата с наибольшем количеством появлений:", key_with_max_value)
print("Количество появлений:", count_dict[key_with_max_value])
print('Ответ 3')
````

#### Task 8.
1. Найдите максимальную разницу между медианами суммы транзакций, посчитанными при заданных ниже условиях по полю amount из таблицы transactions (\*):
    - Медиана суммы транзакций
    - Медиана суммы транзакций по тем строкам, которые ни в одном из своих столбцов не содержат пустые значения
    - Медиана суммы транзакций по строкам, отсортированным по полю amount в порядке возрастания, и из которых удалены дублирующиеся по столбцам [mcc_code, tr_type] строки, причём при удалении соответстующих дублей остаются только последние из дублирующихся строк (keep='last')

Выведите ответ в виде вещественного числа, округлённого до двух знаков после запятой, отделив дробную часть точкой в формате "123.45"

##### Пояснения:
(\*) Для вычисления максимальной разницы между значениями списка можно использовать функцию np.ptp  
(\**) Если в результате получились значения [1,3,5], то макимальная разница между ними 4 == 5-1.

````python
transactions = pd.read_csv('data/transactions.csv', sep=',')



# Вычисление медиан для трех условий
median_total = transactions['amount'].median()
median_no_missing = transactions['amount'].dropna().median()

sorted_unique_rows = transactions.sort_values(by='amount', ascending=True).drop_duplicates(
    subset=['mcc_code', 'tr_type'], keep='last')
median_sorted_unique = sorted_unique_rows['amount'].median()

# Вычисление максимальной разницы между медианами
max_median_difference = np.ptp([median_total, median_no_missing, median_sorted_unique])

print(f"Медиана суммы транзакций: {median_total:.2f}")
print(f"Медиана суммы транзакций без пропущенных значений: {median_no_missing:.2f}")
print(f"Медиана суммы транзакций для уникальных строк: {median_sorted_unique:.2f}")
print(f"Максимальная разница между медианами: {max_median_difference:.2f}")
````
